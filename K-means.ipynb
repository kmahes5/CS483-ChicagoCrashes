```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
```

```python
data = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
locations = data[['LATITUDE','LONGITUDE']]
locations = locations.dropna()
locations = locations[locations.LATITUDE != 0] #this was the issue plaguing the plot smh
```

```python
# Building the model 
kmeans = KMeans(n_clusters = 10) 
 
# Training the model 
kmeans.fit(locations) 
 
# Storing the predicted Clustering labels 
labels = kmeans.predict(locations) 
centers = np.array(kmeans.cluster_centers_)
```

```python
plt.scatter(locations['LONGITUDE'], locations['LATITUDE'], c = labels,)
plt.scatter(centers[:,1], centers[:,0], marker='*', color='k', s=150)
plt.title("Center points of crashes in Chicago determined by K-Means")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```

```python
kmeans2 = KMeans(n_clusters = 15) 
 
kmeans2.fit(locations) 
labels2 = kmeans2.predict(locations) 
centers2 = np.array(kmeans2.cluster_centers_)

plt.scatter(locations['LONGITUDE'], locations['LATITUDE'], c = labels2,)
plt.scatter(centers2[:,1], centers2[:,0], marker='*', color='k', s=150)
plt.title("Center points of crashes in Chicago determined by K-Means")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```
